````python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate dates for 24 months of data
dates = pd.date_range(start='2023-01-01', end='2024-12-31', freq='M')

# Create base data
data = {
    'Date': dates,
    'System_Type': np.random.choice(['Perimeter_Defense', 'Drone_Detection', 'Target_Acquisition', 'Threat_Assessment'], size=len(dates)),
    'Deployment_Location': np.random.choice(['Location_A', 'Location_B', 'Location_C', 'Location_D'], size=len(dates)),
    'Contract_Value_MM': np.random.uniform(5, 50, size=len(dates)),
    'AI_Detection_Accuracy': np.random.uniform(0.92, 0.99, size=len(dates)),
    'False_Positive_Rate': np.random.uniform(0.01, 0.05, size=len(dates)),
    'Response_Time_MS': np.random.uniform(50, 150, size=len(dates)),
    'System_Uptime_Percent': np.random.uniform(0.985, 0.999, size=len(dates)),
    'Maintenance_Hours': np.random.uniform(10, 40, size=len(dates)),
    'Customer_Satisfaction': np.random.uniform(4.0, 5.0, size=len(dates)),
    'Cost_Savings_Percent': np.random.uniform(0.15, 0.35, size=len(dates))
}

# Create DataFrame
df = pd.DataFrame(data)

# Add some calculated fields
df['ROI'] = (df['Contract_Value_MM'] * (1 + df['Cost_Savings_Percent'])) / df['Contract_Value_MM']
df['Operational_Score'] = (df['AI_Detection_Accuracy'] * 0.4 + 
                          (1 - df['False_Positive_Rate']) * 0.3 + 
                          df['System_Uptime_Percent'] * 0.3)

# Add quarterly targets and achievements
df['Revenue_Target_MM'] = np.where(df.index % 3 == 0, 
                                  df['Contract_Value_MM'].rolling(3).sum(), 
                                  np.nan)
df['Revenue_Achievement_Percent'] = np.where(df.index % 3 == 0,
                                           (df['Contract_Value_MM'].rolling(3).sum() / 
                                            df['Revenue_Target_MM'] * 100),
                                           np.nan)

# Print first few rows and basic statistics
print("\nSample Data:")
print(df.head())
print("\nSummary Statistics:")
print(df.describe())
````
